```python
#step2:key generate
key = input()
def to_binary(plantext):
    hex_string = plantext.encode('utf-8').hex()
    print(f'>>> the hexa representation of plantext is: {hex_string}')
    decimal_number = int(hex_string, 16)  # Convert hexadecimal to decimal
    binary_string = bin(decimal_number)[2:]  # Convert decimal to binary and remove '0b' prefix
    while len(binary_string) < len(hex_string) * 4:
        binary_string = '0' + binary_string
    return binary_string
key_binary_code = to_binary(key)
print(f'>>> key to binary code: {key_binary_code}')

table_PC1 = [
    57, 49, 41, 33, 25, 17, 9,
    1, 58, 50, 42, 34, 26, 18,
    10, 2, 59, 51, 43, 35, 27,
    19, 11, 3, 60, 52, 44, 36,
    63, 55, 47, 39, 31, 23, 15, 
    7, 62, 54, 46, 38, 30, 22,
    14, 6, 61, 53, 45, 37, 29, 
    21, 13, 5, 28, 20, 12, 4
]
def _56bit_permutation_key(key):
    permutation_key = ''
    for x in table_PC1:
        permutation_key += key[x - 1]
    return permutation_key
permutation_key = _56bit_permutation_key(key_binary_code)
print(f'>>> permutation key: {permutation_key} {len(permutation_key)}')

def shift_1(list_binary):
    first_binary = list_binary[0]
    list_binary = list_binary[1:]
    list_binary += first_binary
    return list_binary
    
def shift_2(list_binary):
    first_binary = list_binary[0]
    second_binary = list_binary[1]
    list_binary = list_binary[2:]
    list_binary += first_binary
    list_binary += second_binary
    return list_binary
table_PC2 = [
    14, 17, 11, 24, 1, 5,
    3, 28, 15, 6, 21, 10,
    23, 19, 12, 4, 26, 8,
    16, 7, 27, 20, 13, 2,
    41, 52, 31, 37, 47, 55,
    30, 40, 51, 45, 33, 48,
    44, 49, 39, 56, 34, 53,
    46, 42, 50, 36, 29, 32
]
_16subkeys = []
_CD_pairs = []
_CD_pairs.append('11110000110011001010101011110101010101100110011110001111')

def generate_key(original_key):
    i = 0
    while i < 17:
        CD_pair_value = ''
        c = _CD_pairs[i][:28]
        d = _CD_pairs[i][28:]
        print(f'>>> C{i}: {c}')
        print(f'>>> D{i}: {d}')
        if i == 0 or i == 1 or i == 8 or i == 15:
            CD_pair_value += shift_1(c)
            CD_pair_value += shift_1(d)
        else:
            CD_pair_value += shift_2(c)
            CD_pair_value += shift_2(d)
        key_current = ''
        for x in table_PC2:
            key_current += CD_pair_value[x - 1]
        _CD_pairs.append(CD_pair_value)
        _16subkeys.append(key_current)
        i += 1
generate_key('11110000110011001010101011110101010101100110011110001111')
_16subkeys = _16subkeys[:16]
for i in range(len(_16subkeys)):
    print(f'>>> K{i+1}: {_16subkeys[i]}')
```

<p style="padding-left: 40px">- Whitener (XOR):
sau khi mở rộng phép hoán vị, DES sử dụng phép XOR

<p style="padding-left: 40px">-Hàm F:</p>
<p style="padding-left: 40px">Đầu vào hàm f có 2 biến:</p>
<p style="padding-left: 60px">+ Biến thứ nhất: Ri-1 là xâu bit có độ dài 32 bit. - Biến thứ hai: Ki là xâu bit có độ dài 48 bit. Đầu ra của hàm f là xâu có độ dài 32 bit. Quy trình hoạt động của hàm f như sau: </p>
<p style="padding-left: 60px">+ Biến thứ nhất Ri-1 được mở rộng thành một xâu có độ dài 48 bit theo một hàm mở rộng hoán vị E (Expansion permutation). Thực chất hàm mở rộng E(Ri-1) là một hoán vị có lặp trong đó lặp lại 16 bit của Ri-1.</p>
<p style="padding-left: 60px">+ Tính E(Ri-1) XOR Ki. </p>
<p style="padding-left: 60px">+ Tách kết quả của phép tính trên thành 8 xâu 6 bit B1, B2, …, B8.</p>
<p style="padding-left: 60px">+ Đưa các khối 8 bit Bi vào 8 bảng S1, S2, …, S8 (được gọi là các hộp S-box). Mỗi hộp S-Box là một bảng 4*16 cố định có các cột từ 0 đến 15 và các hàng từ 0 đến 3. Với mỗi xâu 6 bit Bi = b1b2b3b4b5b6 ta tính được SiBi như sau: hai bit b1b6 xác định hàng r trong hộp Si, bốn bit b2b3b4b5 xác định cột c trong hộp Si. Khi đó, Si(Bi) sẽ xác định phần tử Ci = Si(r,c), phần tử này viết dưới dạng nhị phân 4 bit. Như vậy, 8 khối 6 bit Bi (1 ≤ i ≤ 8) sẽ cho ra 8 khối 4 bit Ci với (1 ≤ i ≤ 8). </p>
<p style="padding-left: 60px">+ Xâu bit C = C1C2C3C4C5C6C7C8 có độ dài 32 bit được hoán vị theo phép toán hoán vị P (hộp P-Box). Kết quả P(C) sẽ là kết quá của hàm f(Ri-1,Ki).</p>

```python
#step3:
plantext_initial = binary_number
def initial_permutation_IP(plantext):
    IP = ''
    for x in initial_permutation_table:
        IP += plantext[x - 1]
    return IP
L0R0 = initial_permutation_IP(plantext_initial)
# print(f'>>> plantext: {binary_number}')
# print(f'>>> initial permutation: {initial_permutation_IP(binary_number)}')


def expansion(binary):
    binary_expanded = ''
    for x in expansion_table:
        binary_expanded += binary[x - 1]
    return binary_expanded
# print(expansion(L0R0[32:]))
# print(len(expansion(L0R0[32:])))

#XOR_operator
def XOR_operator(a, b):
    result = ''
    for x in range(len(a)):
        if (a[x] == '0' and b[x] == '0') or (a[x] == '1' and b[x] == '1'):
            result += '0'
        else:
            result += '1'
    return result
#convert int to binary
def int_to_binary_4bit(n):
    # Convert integer to binary representation
    binary_string = bin(n)[2:]  # Remove '0b' prefix
    
    # Add leading zeros to ensure 4-bit representation
    while len(binary_string) < 4:
        binary_string = '0' + binary_string
    
    return binary_string
#s_function
def s_function(b,index):
    row = int(b[0] + b[5],2)
    column = int(b[1:5],2)
    return int_to_binary_4bit(s_boxes[index][row][column])

#f function
def f_function(r, key):
    f = ''
    result = ''
    RK = XOR_operator(r, key)
    i, index = 0, 0
    B = ''
    for x in range(len(RK)):
        i += 1
        B += RK[x]
        if i == 6:
            result += s_function(B, index)
            index +=1
            i = 0
            B = ''
    for x in permutation_p_table:
        f += result[x - 1]
    return f

#16round and final permutation 
# def final_permutation(array):
#     result = ''
#     for x in final_permutation_table:
#         result += array[x]
#     return result
array_LR = []
def _16round():
    R_current = L0R0[32:]
    L_current = L0R0[:32]
    print(f'>>> L0: {L_current}')
    print(f'>>> R0: {R_current}')
    i = 0
    while i < 16:
        L = R_current
        R_current = expansion(R_current)
        R = XOR_operator(L_current, f_function(R_current, _16subkeys[i]))
#         print(f'>>> L{i+1}: {L}')
#         print(f'>>> R{i+1}: {R}')
        i += 1
        L_current = L
        R_current = R
        array_LR.append(L_current+R_current)
        if i == 16:
            return R_current + L_current

L16R16 = _16round()
for i in range(len(array_LR)):
    print(f'>>> L{i+1}: {array_LR[i][:32]}')
    print(f'>>> R{i+1}: {array_LR[i][32:]}')
print(f'>>> R16L16: {L16R16}')

#Ip CONVERT
cyphertext_binary = ''
for x in final_permutation_table:
    cyphertext_binary += L16R16[x - 1]
# hexadecimal_number = hex(int(binary_number, 2))
print(f'>>> cyphertext binary: {cyphertext_binary}')
cyphertext = hex(int(cyphertext_binary,2))
print(f'>>> cyphertext: {cyphertext}')
```
